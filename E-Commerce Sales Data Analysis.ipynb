```python
import pandas as pd
import matplotlib.pyplot as plt

data = {
    'Month': pd.date_range(start='2021-01-01', periods=36, freq='M'),
    'Sales': [120, 135, 150, 145, 160, 155, 180, 175, 190, 185, 200, 195,
              210, 205, 220, 215, 230, 225, 240, 235, 250, 245, 260, 255,
              270, 275, 280, 285, 290, 295, 300, 305, 310, 315, 320, 325]
}

df = pd.DataFrame(data)

df['Year'] = df['Month'].dt.year
annual_sales = df.groupby('Year')['Sales'].sum().reset_index()

plt.figure(figsize=(10, 5))
plt.plot(annual_sales['Year'], annual_sales['Sales'], marker='o')
plt.xlabel('Year')
plt.ylabel('Total Sales')
plt.title('Annual Sales Performance Over Time')
plt.grid(True)
plt.show()
```

```python
# 2. The client is interested in understanding how the sales have been fluctuating 
#     over time to identify any underlying patterns or anomalies Analyze the trend in
#     our monthly sales using a moving average technique.
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

data = {
    'Month': pd.date_range(start='2022-01-01', periods=24, freq='M'),
    'Sales': [120, 135, 150, 145, 160, 155, 180, 175, 190, 185, 200, 195,
              210, 205, 220, 215, 230, 225, 240, 235, 250, 245, 260, 255]
}

df = pd.DataFrame(data)

df.head()
```

```python
df['3_Month_MA'] = df['Sales'].rolling(window=3).mean()

df.head()
```

```python
plt.figure(figsize=(20,10))
plt.plot(df['Month'], df['Sales'], label='Monthly Sales')
plt.plot(df['Month'], df['3_Month_MA'], label='3-Month Moving Average', color='orange')
plt.xlabel('Month')
plt.ylabel('Sales')
plt.title('Monthly Sales and 3-Month Moving Average')
plt.xticks(rotation=20)
plt.legend()
plt.grid(True)
plt.show()


```

```python
import pandas as pd
import matplotlib.pyplot as plt

data = {
    'Month': pd.date_range(start='2021-01-01', periods=36, freq='M'),
    'Sales': [120, 135, 150, 145, 160, 155, 180, 175, 190, 185, 200, 195,
              210, 205, 220, 215, 230, 225, 240, 235, 250, 245, 260, 255,
              270, 275, 280, 285, 290, 295, 300, 305, 310, 315, 320, 325]
}

df = pd.DataFrame(data)

df['MoM_Growth_Rate'] = df['Sales'].pct_change() * 100

plt.figure(figsize=(12, 6))
plt.plot(df['Month'], df['MoM_Growth_Rate'], marker='o', linestyle='-')
plt.xlabel('Month')
plt.ylabel('MoM Growth Rate (%)')
plt.title('Month-Over-Month Growth Rate in Sales')
plt.grid(True)
plt.axhline(0, color='red', linestyle='--')
plt.show()
```

```python
# 5. What are the total sales made by each employee?
```

```python
import pandas as pd
import matplotlib.pyplot as plt

data = {
    'Representative': ['Anoop', 'Sumit', 'Anupam', 'Anoop', 'Sumit', 'Anupam', 'Anoop', 'Sumit', 'Anupam'],
    'Sales': [500, 700, 400, 600, 800, 450, 700, 850, 500]
}

df = pd.DataFrame(data)

total_sales_by_rep = df.groupby('Representative')['Sales'].sum().reset_index()

top_performers = total_sales_by_rep.sort_values(by='Sales', ascending=False)

# Display the top performers
print(top_performers)

plt.figure(figsize=(14, 7))
plt.bar(top_performers['Representative'], top_performers['Sales'], color='pink')
plt.xlabel('Representative')
plt.ylabel('Total Sales')
plt.title('Top-Performing Sales Representatives')
plt.xticks(rotation=20)
plt.grid(axis='y')
plt.show()
```

```python
# 13. Which subcategory has the highest average unit price?
```

```python
import pandas as pd

data = {
    'Subcategory': ['Electronics', 'Clothing', 'Electronics', 'Furniture', 'Clothing', 'Furniture'],
    'Unit Price': [100, 80, 120, 150, 90, 180]
}

df = pd.DataFrame(data)

average_unit_price_by_subcategory = df.groupby('Subcategory')['Unit Price'].mean()

highest_avg_unit_price_subcategory = average_unit_price_by_subcategory.idxmax()
highest_avg_unit_price = average_unit_price_by_subcategory.max()

print(f"The subcategory with the highest average unit price is {highest_avg_unit_price_subcategory} with an average unit price of {highest_avg_unit_price}.")
```
