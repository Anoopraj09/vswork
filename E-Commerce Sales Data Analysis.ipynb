```python
import matplotlib.pyplot as plt

# Sample data: replace these lists with your actual data
years = [2015, 2016, 2017, 2018, 2019, 2020, 2021, 2022, 2023]
sales = [25000, 27000, 30000, 32000, 35000, 37000, 40000, 42000, 45000]

plt.figure(figsize=(10, 5))
plt.plot(years, sales, marker='o', linestyle='-', color='b')

plt.title('Annual Sales Performance Over Time')
plt.xlabel('Year')
plt.ylabel('Sales (in USD)')
plt.grid(True)
plt.xticks(years)
plt.tight_layout()

# Save the plot as an image file
plt.savefig('annual_sales_performance.png')

# Show the plot
plt.show()
```
